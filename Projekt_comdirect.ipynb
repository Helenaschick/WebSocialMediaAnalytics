```python
#Libraries importieren
import pandas as pd
from bs4 import BeautifulSoup
import requests
import lxml 
```

```python
#Zugriff auf comdirekt Webseite
r = requests.get('https://www.comdirect.de/inf/maerkte/zinsuebersicht.html').content
soup = BeautifulSoup(r, 'lxml')
```

```python
#Zinssatz suchen (mit find wird erster Fund genommen)
zins = soup.find("td", {"class": "table__column table__column--right"}).text
print("Euro Short-Term Rate:", zins)
```

```python
#alle Zinssätze (aktuell bis 1 Monat, class für 1 Jahr ist anders)
zins = soup.find_all("td", {"class": "table__column table__column--right"})
zins
```

```python
#alle Zinsen Tagesgeld USA
zins_aktuell = soup.find("td", {"class": "table__column table__column--right"}).text
zins_1jahr = soup.find("td", {"class": "table__column table__column--right outer-spacing--"}).text

#zins_1tag = soup.find("td", {"class": "table__column table__column--right"}, {"data-label": "1 Tag"}).text #nimmt aktuell noch den aktuellen Wert, nicht Tag 1
#zins_1monat = soup.find("td", {"class": "table__column table__column--right"}, {"data-label": "1 Monat"}).text #nimmt aktuell noch den aktuellen wert, nicht Monat 1
```

```python
print("Zins aktuell:", zins_aktuell)
#print("Zins 1 Tag:", zins_1tag)
#print("Zins 1 Monat:", zins_1monat)
print("Zins 1 Jahr:", zins_1jahr)
```
